```python
# Fonction donnant la distribution empirique d'une variable catégorielle ou discrète avec un petit nombre de modalités réalisées
def plot_cat(variable, order=None, rotation=None):
    prct_df = (df[variable]
           .value_counts(normalize=True)
           .mul(100)
           .rename("Fréquence (%)")
           .reset_index())
    sns.barplot(data=prct_df, x="index", y="Fréquence (%)", color='tab:blue', order=order).set(xlabel=variable)
    plt.grid(axis="y")
    if rotation is not None:
        plt.xticks(rotation=rotation)
    pdf.export()
    plt.show()
```

```python
# Fonction donnant la distribution empirique d'une variable continue ou discrète avec un grand nombre de modalités réalisées
def plot_quant(variable, binwidth=1, quantile_max=None):
    data = df[df[variable] < df[variable].quantile(q=0.9)] if quantile_max is not None else df
    xlabel = f"{variable} < q_{quantile_max}" if quantile_max is not None else variable
    sns.histplot(data=data, x=variable, binwidth=binwidth, stat="density").set(xlabel=xlabel, ylabel="Densité")
    plt.grid()
    pdf.export()
    plt.show()
```
